# Matplotlib 

### Подключение необходимых для работы библиотек 

```python
import matplotlib.pyplot as plt
import warnings
import numpy as np
warnings.filterwarnings('ignore')
```

### Генерируем данные

```python
np.random.seed(1)
x = np.arange(10)
y1 = np.random.randint(10, size=10)
y2 = np.random.randint(10, size=10)
```

### Рисуем линию по значениям x,y1 и точки по значениям x,y2

```python
plt.plot(x, y1)
plt.scatter(x, y2,color='red')
plt.show()
```

### Рисуем столбцы по данным x,y1 

```python
plt.bar(x, y1)
plt.show()
```

### Рисуем столбцы по данным x,y1 и x,y2 на одном графике 

```python
plt.bar(x, y1)
plt.bar(x, y2, bottom=y1)
plt.show()
```

### Гистограмма распределения данных 

```python
vals = np.random.rand(100)
plt.hist(vals, bins=10)
plt.show()
```

### Работа с графиками

```python
plt.plot(x, y1)
# Подписи осей
plt.xlabel('Время')
plt.ylabel('Значение')
# Заголовок
plt.title('График №1')
# Включение сетки
plt.grid()
plt.show()
```

### Подпись делений по оси х

```python
x_mod = np.char.add(x.astype(str), ' сек')

plt.plot(x, y1)
plt.xlabel('Время')
plt.xticks(x, x_mod)
plt.grid()
plt.show()
```

### Размер графика

```python
x_mod = np.char.add(x.astype(str), ' сек')

plt.figure(figsize=(8, 4))
plt.plot(x, y1)
plt.xlabel('Время')
plt.xticks(x, x_mod)
plt.grid()
plt.show()

```

### Добавление легенды

```python
x_mod = np.char.add(x.astype(str), ' сек')

plt.figure(figsize=(8, 4))
plt.plot(x, y1, label = "Данные 1")
plt.plot(x, y2, label = "Данные 2")
plt.xlabel('Время')
plt.xticks(x, x_mod)
plt.legend()
plt.grid()
plt.show()
```

### Вывод нескольких графиков горизонтально

```python
x_mod = np.char.add(x.astype(str), ' сек')

plt.figure(figsize=(14, 4))


plt.subplot(1, 2, 1)
plt.plot(x, y1, label = "Данные 1")
plt.xlabel('Время')
plt.xticks(x, x_mod)
plt.legend()
plt.grid()


plt.subplot(1, 2, 2)
plt.plot(x, y2, label = "Данные 2", color = 'r')
plt.xlabel('Время')
plt.xticks(x, x_mod)
plt.legend()
plt.grid()

plt.show()
```

### Вывод нескольких графиков вертикально 

```python
x_mod = np.char.add(x.astype(str), ' сек')

plt.figure(figsize=(8, 8))


plt.subplot(2, 1, 1)
plt.plot(x, y1, label = "Данные 1")
plt.xlabel('Время')
plt.xticks(x, x_mod)
plt.legend()
plt.grid()


plt.subplot(2, 1, 2)
plt.plot(x, y2, label = "Данные 2", color = 'r')
plt.xlabel('Время')
plt.xticks(x, x_mod)
plt.legend()
plt.grid()

plt.show()
```

### Параметры кривой 

```python
x_mod = np.char.add(x.astype(str), ' сек')

plt.figure(figsize=(8, 4))

plt.plot(x, y1, '-.', label = "Данные 1", color = 'green', linewidth = 5)
plt.xlabel('Время')
plt.xticks(x, x_mod)
plt.legend()
plt.grid()
```

### Параметры шрифта

```python
csfont = {'fontname':'Comic Sans MS'}


plt.figure(figsize=(8, 4))

plt.plot(x, y1, '-.', label = "Данные 1", color = 'green', linewidth = 5)
plt.xlabel('Время', **csfont)
plt.xticks(x, x_mod)
plt.title('title',**csfont)
plt.legend()
plt.grid()
```

```python
plt.rcParams['font.sans-serif'] = "Comic Sans MS"
plt.rcParams['font.family'] = "sans-serif"

plt.rcParams.update({'font.size': 14})

plt.figure(figsize=(8, 4))

plt.plot(x, y1, '-.', label = "Данные 1", color = 'green', linewidth = 5)
plt.xlabel('Время')
plt.xticks(x, x_mod)
plt.legend()
plt.grid()
```
